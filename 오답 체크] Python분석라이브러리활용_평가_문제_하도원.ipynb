### Q11. age를 7개 카테고리로 구분하는 새로운 칼럼 'cat_age'를 생성하여 출력하세요. 단, 카테고리 구분을 수행하는 사용자 함수를 만들고 그 함수를 age 칼럼에 매핑하여 결과를 tdf1에 저장하고 출력하세요.

[카테고리]  
age <= 5: cat = 'Baby'  
age <= 12: cat = 'Child'  
age <= 18: cat = 'Teenager'  
age <= 25: cat = 'Student'  
age <= 60: cat = 'Adult'  
age > 60 : cat = 'Elderly'

```python
# A.
def cat_age(x): #@ 사용자 함수를 만드네. 이걸 언제 만들어야 할지 잘 생각해보자.
    cat = ''
    if   x <= 5   :  cat = 'Baby'
    elif x <= 12  :  cat = 'Child'
    elif x <= 18  :  cat = 'Teenager'
    elif x <= 25  :  cat = 'Student'
    elif x <= 60  :  cat = 'Adult'
    else : cat = 'Elderly' #  x 없이 : 을 붙여야 한다. -> else: 형식이니까 그렇지 뭐. if: 듯이. 한 칸 아래로 내리기가 익숙하지 않아서 생각을 좀 해야 했다.
        
    return cat

tdf1 = tdf.copy()
tdf1['age_cat'] = tdf.age.apply(cat_age)

tdf1.head()
```

```python
tdf1 = tdf.agg({'cat_age':['survived','sex','age','class']})
tdf1
```

### Q13. join() 메소드는 두 데이터프레임의 행 인덱스를 기준으로 결합한다. 2개의 주식데이터를 가져와서 join() 메소드로 아래와 같이 결합한 후 다음 사항을 수행하세요.
- df1과 df2의 교집합만 출력되도록 결합하여 df4에 저장하고 출력
- df4에서 중복된 칼럼을 삭제한 후 블린 인덱싱을 이용하여 eps가 3000 보다 적거나 stock_name이 이마트인 데이터를 선택하여 데이터프레임을 생성하고 df5 이름으로 저장 및 출력하세요.(단, '<' 와 '==' 를 반드시 사용해야 함)

```python
# A.

df1 = pd.read_excel('C:/workspace/cakd3/과제 + 시험/stock price.xlsx', index_col='id')
df2 = pd.read_excel('C:/workspace/cakd3/과제 + 시험/stock valuation.xlsx', index_col='id')
df4 = df1.join(df2, how='inner') # df1과 df2의 교집합만 출력되도록 결합하여 df4에
# how='inner' 는 join( 데이터프레임 명 , ) 의 디폴트 값이다.
# - inner join 을 A 와 B에 대해 수행하는 것은, A와 B의 교집합을 말합니다. 벤다이어그램으로 그렸을 때 교차되는 부분입니다.
# - outer join을 A와 B에 대해 수행하는 것은, A와 B의 합집합을 말합니다. 벤다이어 그램으로 그렸을 때, 합집합 부분입니다.
# 출처: https://stanleykou.tistory.com/entry/SQL-INNER-조인과-OUTER조인이-무엇인가요 [StanleyKou의 개인작업실]
print(df4)
print()
df4.drop('name', axis=1, inplace=True) # stock_name과 name은 서로 중복되는 값이므로 칼럼 삭제.
# inplace=True => 값을 계산한 후 사본을 삭제


df5 = df4[(df4['eps'] < 3000) | (df4['stock_name'] == '이마트')] # 이거나. 합집함을 선택 해야겠지.
df5
```

1. Pandas는 원본 데이터의 복사본을 만듭니다.

### Q15. 'mpg'를 'kpl' 로 환산하여 새로운 열을 생성하고 처음 3개 행을 소수점 아래 둘째 자리에서 반올림하여 출력하세요.

### Q17. titanic 데이터셋에서 5개 열을 선택해서 class열을 기준으로 그룹화를 수행한 후 아래와 같이 출력하였다. 다음 사항을 출력하세요.

- 그룹별 평균 출력
- 그룹별 최대값 출력

```python
# A.
import pandas as pd
import seaborn as sns

# titanic 데이터셋에서 age, sex 등 5개 열을 선택하여 데이터프레임 만들기
titanic = sns.load_dataset('titanic') # sns.load_dataset('파일명') 으로 불러오네.
                                      # C:\Users\Ha dowon\seaborn-data 에 있다. seaborn은 seaborn-date 파일을 따로 둔다.
df = titanic.loc[:, ['age','sex', 'class', 'fare', 'survived']] # 5개 열을 선택해서 df(데이터프레임)로 만들었네.

print(df.head())
print()
# class 열을 기준으로 그룹화(1개 열을 기준)
grouped = df.groupby(['class']) # .groupby() 안의 인자를 기준으로 그룹화 수행하는 함수 -> 안에 인자는 ['열 이름']
print('그룹별 평균')
average = grouped.mean() # 그룹별 평균
print(average)
print()
print('그룹별 최대값')
stat = grouped.max() # 그룹별 최대값
print(stat)
```

### Q20.  titanic 데이터셋에서 다음 전처리를 수행하세요.
1. df에서 중복 칼럼으로 고려할 수 있는 컬럼들(6개 내외)을 삭제한 후 나머지 컬럼들로 구성되는 데이터프레임을 df1 이름으로 저장 후 출력하세요. 
2. df1에서 null값이 50% 이상인 칼럼을 삭제 후 df2 이름으로 저장하고 출력하세요.
3. df2에서 결측값이 있는 age 칼럼에 대해서 평균값으로 대체 처리를 수행하세요. 
4. df2에서 결측값이 있는 embarked 칼럼에 대해서 앞행의 값으로 대체 처리를 수행하세요. 
5. df2 문자로 되어있는 칼럼들을 레이블 인코딩 수행하여 숫자로 변환 후 df2.info()를 출력하세요

```python
import seaborn as sns
df = sns.load_dataset('titanic')
df.head()
```

```python
# 1. df에서 중복 칼럼으로 고려할 수 있는 컬럼들(6개 내외)을 삭제한 후 
# 나머지 컬럼들로 구성되는 데이터프레임을 df1 이름으로 저장 후 출력하세요. 

# A-1.
df.columns
df1 = df[['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'fare',
       'embarked', 'deck']] # df에서 중복 칼럼 고려하여 삭제한다.
df1.head()
```

```python
# 2. 풀이에 필요
df1.isnull().sum()
```

```python
# 2. df1에서 null값이 50% 이상인 칼럼을 삭제 후 
# df2 이름으로 저장하고 출력하세요.

# A-2.
df2 = df1.drop('deck',axis=1) # 'deck' 은 df1에서 null값이 50% 이상인 칼럼이다. -> 삭제
df2.head()
```

```python
# 3. df2에서 결측값이 있는 age 칼럼에 대해서 평균값으로 대체 처리를 수행하세요. 
## 결측값 은 null 이지.
## 대체처리가 뭐지?  -> ,inplace=True
## copy해서 대체 처리한 후 삭제하니까. 대체처리라고 부를 수 있다.
# A-3.
df2.age.isnull().sum()
# 위 구문 결괏값은 177 나온다.

df2.age.fillna(df2.age.mean(),inplace=True) # 여기서 평균값으로 대체 처리를 수행
df2.age.isnull().sum() # 그리고 다시 결측값을 확인해본다.
```

```python
# 4. df2에서 결측값이 있는 embarked 칼럼에 대해서 앞행의 값으로 대체 처리를 수행하세요. 
## embarked 칼럼에도 결측값이 2개 있다.

## 앞 행의 값으로 대체 처리를 하라는 게 무슨 말이지? 앞 행은 fare 인데?
## -> 
# A-4.
df2.embarked.isnull().sum()
df2.embarked.fillna(method='ffill',inplace=True) #@ method='ffill' 가 뭐지?
df2.embarked.isnull().sum()
```

```python
# 5. df2 문자로 되어있는 칼럼들을 레이블 인코딩 수행하여 숫자로 변환 후 
# df2.info()를 출력하세요

## sklearn 에서 preprocessing 을 import 한다.

# A5.
from sklearn import preprocessing

features = ['sex','embarked'] # 문자로 되어 있는 칼럼들
for feature in features: # features 변수의 칼럼들에서 하나하나 뽑아낸다. 인코딩 하려고.
    le = preprocessing.LabelEncoder() # preprocessing.LabelEncoder() -> 인코딩 수행
    df2[feature] = le.fit_transform(df2[feature]) # .fit_transform() 으로 숫자로 변환한다.
            # df2의 feature에 해당하는 걸 숫자로 변환. feature은 sex 컬럼과 embarked 칼럼에서 하나하나 뽑아낸 거지.
df2.info()
```
